```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_json('LiveWeatherData.json')
df.head(2)
```

```python
data = df[['City', 'Air_Quality']]
data.head()
```

```python
# Line Plot For Data Visualization
df = pd.DataFrame(data)
df.plot(x='City', y='Air_Quality', kind='line', marker='o')
plt.xlabel('City')
plt.ylabel('Air_Quality')
plt.title('Weather data')
plt.show()
```

```python
car = ["Caterham", "Tesla", "Audi", "BMW", "Ford", "Jeep"]
weight = [0.48, 1.7, 2, 2, 2.3, 3]

# create a DataFrame
data = {'Car': car, 'Weight': weight}
df = pd.DataFrame(data)
df.plot(x='Car', y='Weight', kind='line', marker='o')
plt.xlabel('Car')
plt.ylabel('Weight')
plt.title("Cars_Weight")
plt.show()
```

```python
# Scatter Plots For Data Visualization
std_data = {'Name':['Aqib','Shami','Zubi','Ahmad'],
            'CGPA':[3.34,3.39,3.29,3.56]}
df = pd.DataFrame(std_data)
df.plot(x='Name', y='CGPA', kind='scatter', marker='o', color='blue')
plt.xlabel('Names')
plt.ylabel('CGPA')
plt.title('Student_Marks')
plt.show()
```

```python
df.plot(x='Name',y='CGPA', kind='bar', color='green')
plt.xlabel('Names')
plt.ylabel('CGPA')
plt.tight_layout()
plt.title('Marks')
plt.show()
```

```python
df['CGPA'].plot(kind='hist',bins=10, edgecolor='white', color='green')
plt.show()
```

```python
# Plot a histogram 
data = {'values': [18, 21, 22, 25, 28, 30, 32, 34, 35, 36, 38, 40, 42, 45, 50, 55, 60, 65, 70]}
df = pd.DataFrame(data)
plt.hist(df['values'], bins=10)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data1 = {'values': [12, 15, 18, 22, 25, 27, 30, 33, 37, 40]}
data2 = {'values': [8, 10, 14, 20, 24, 28, 32, 36, 42, 45]}
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

plt.hist(df1['values'], bins=6, edgecolor='black', alpha=0.7, label='Data1')
plt.hist(df2['values'], bins=6, edgecolor='black', alpha=0.7, label='Data2', color='orange')
plt.title('Comparison of two histogram')
plt.xlabel('values')
plt.ylabel('frequency')
plt.legend()
plt.show()
```
